**Feature Engineering**

```python
import pandas as pd

# Convert 'Formatted Date' to datetime, handling errors
data_df['Formatted Date'] = pd.to_datetime(data_df['Formatted Date'], errors='coerce')

# Drop rows with invalid dates
data_df = data_df.dropna(subset=['Formatted Date'])

# Ensure 'Formatted Date' is of datetime type before using .dt accessor
if pd.api.types.is_datetime64_any_dtype(data_df['Formatted Date']):
    data_df['Year'] = data_df['Formatted Date'].dt.year
    data_df['Month'] = data_df['Formatted Date'].dt.month
    data_df['Day'] = data_df['Formatted Date'].dt.day
    data_df['Hour'] = data_df['Formatted Date'].dt.hour
    data_df['DayOfWeek'] = data_df['Formatted Date'].dt.dayofweek
else:
    print("Error: 'Formatted Date' column is not of datetime type.")
    # Further investigate and fix the data type of 'Formatted Date' column
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as mn
```

** One Hot Encoding and Label Encoding **

```python
# Display column names to confirm one-hot encoding and scaling steps
print(data_df.columns)
```

```python
### Step 1: One-Hot Encoding for Categorical Data
# Apply one-hot encoding to 'Summary' and 'Precip Type' (or any other categorical columns found)
categorical_cols = ['Summary', 'Precip Type']  # Update if you have additional categorical columns

# Get dummy variables for categorical features
data_df = pd.get_dummies(data_df, columns=categorical_cols, drop_first=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# Example of Label Encoding for an ordinal categorical variable
if 'Wind Direction' in data_df.columns:
    le = LabelEncoder()
    data_df['Wind Direction'] = le.fit_transform(data_df['Wind Direction'])
```

**Data Scaling**

```python
from sklearn.preprocessing import MinMaxScaler

# Define feature matrix X again after one-hot encoding
X = data_df.drop(columns=['Temperature (C)', 'Formatted Date'])  # Assuming 'Temperature (C)' is the target
y = data_df['Temperature (C)']

# Select only numerical features for scaling
numerical_features = X.select_dtypes(include=['number']).columns
X_numerical = X[numerical_features]

# MinMax Scaling
minmax_scaler = MinMaxScaler()
X_scaled_minmax = minmax_scaler.fit_transform(X_numerical)

# Create DataFrame from scaled features
X_scaled_minmax_df = pd.DataFrame(X_scaled_minmax, columns=numerical_features, index=X.index)

# Concatenate scaled numerical features with original categorical features
X_final = pd.concat([X_scaled_minmax_df, X.drop(columns=numerical_features)], axis=1)
```

**Train-Test Split**

```python
from sklearn.model_selection import train_test_split

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)

print("Training set size:", X_train.shape)
print("Test set size:", X_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Convert the continuous target to a binary class
y_train_binary = (y_train > y_train.mean()).astype(int)  # 1 for above-average temperature, 0 for below-average
y_test_binary = (y_test > y_test.mean()).astype(int)

# Initialize and fit the Gaussian Naive Bayes model
gnb_model = GaussianNB()
gnb_model.fit(X_train, y_train_binary)

# Make predictions on the test data
y_pred_gnb = gnb_model.predict(X_test)

# Calculate performance metrics
gnb_accuracy = accuracy_score(y_test_binary, y_pred_gnb)
gnb_precision = precision_score(y_test_binary, y_pred_gnb, average='weighted', zero_division=0)
gnb_recall = recall_score(y_test_binary, y_pred_gnb, average='weighted', zero_division=0)
gnb_f1 = f1_score(y_test_binary, y_pred_gnb, average='weighted', zero_division=0)

print("GNB Accuracy:", gnb_accuracy)
print("GNB Precision:", gnb_precision)
print("GNB Recall:", gnb_recall)
print("GNB F1 Score:", gnb_f1)
```
